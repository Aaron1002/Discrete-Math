```python
"""This code can run without problem.

How to run:
1. input the network filename or path
2. it reads the file and stores the graph by adjacency matrix and list
3. it asks the user to input a source node index s, and a data structure type, and then print out adjacent arcs (s,j).
4. it repeats step 3, unless s=0, which terminates the code

This code is written by 王晏國, email h44114025@gs.ncku.edu.tw, on 2023/11/24"""


import re
import numpy as np

# 前處理
def sp_to_txt(sp_file, txt_file):
    with open(sp_file, "r") as sp:
        with open(txt_file, "w") as txt:
            for line in sp:
                if line[0] == "c":
                    continue
                elif line[0] == "t":
                    continue
                elif line[0] == "a":
                    modified_line = re.sub('\s+', ',', line)
                    txt.write(modified_line + '\n')
                elif line[0] == "p":
                    modified_line = re.sub('\s+', ',', line)
                    txt.write(modified_line + '\n')
                elif line[0] == "n":
                    modified_line = re.sub('\s+', ',', line)
                    txt.write(modified_line + '\n')

def extract_numbers(line):
    # 使用正則表達式匹配數字
    numbers = re.findall(r'\d+', line)
    # 將匹配到的數字轉換為整數
    numbers = [int(num) for num in numbers]
    return numbers

def sp_to_list(txt_file):
    in_list = []
    # 開啟.sp檔案以讀取模式
    with open(txt_file, 'r') as num_file:
        for line in num_file:
            # 提取每一行的數字並形成串列
            numbers = extract_numbers(line)
            in_list.append(numbers)

    return in_list

"""-----------------------------------------------------"""

# 產生Matrix
def Adjacency_matrix():
    ad_matrix = np.zeros((numbers_list[1][0], numbers_list[1][0]))  # 生成n*n matrix(初始化為0)
    for index in range(numbers_list[0][1]+2):   # 有400行，加"p"&"n"
        if index <= 1:
            continue
        ad_matrix[numbers_list[index][0]-1][numbers_list[index][1]-1] += 1  # 建立完整matrix

    return ad_matrix

def From_nodes():
    From = np.zeros(numbers_list[0][1])
    for index in range(numbers_list[0][1]):
        From[index] += numbers_list[index+2][0]

    return From

def To_nodes():
    To = np.zeros(numbers_list[0][1])
    for index in range(numbers_list[0][1]):
        To[index] += numbers_list[index + 2][1]

    return To


sp_file = "D:/大學/選修/離散數學/input_100_400_1.sp"
txt_file = "output.txt"
sp_to_txt(sp_file, txt_file)
numbers_list = sp_to_list(txt_file)
# print(numbers_list)
# print(f"Adjacency matrix:\n {Adjacency_matrix()}")
# print(f"From nodes:\n {From_nodes()}")
# print(f"To nodes:\n {To_nodes()}")

# User input
while True:
    source_vertex = int(input("Please input the source vertex index (0 to exit):"))
    if source_vertex == 0:
        break
    if source_vertex > numbers_list[1][0]:
        print("Too many nodes!")
        continue
    elif source_vertex < 0:
        print("The source vertex cannot be negative!")
        continue
    DS = int(input("Please choose (1) Adjacency Matrix (2) Adjacency List:"))
# Results
    if DS == 1:
        ones_counter = 0
        for i in range(numbers_list[1][0]):
            if Adjacency_matrix()[source_vertex-1][i] == 1:
                ones_counter += 1
                print(f"({source_vertex},{i+1})")
        if ones_counter == 0:
            print("No outgoing arcs!")
    elif DS == 2:
        out_arc = 0
        for i in range(numbers_list[0][1]):
            if From_nodes()[i] == source_vertex:
                print(f"({int(From_nodes()[i])},{int(To_nodes()[i])})")
                out_arc += 1
        if out_arc == 0:
            print("No outgoing arcs!")
```
